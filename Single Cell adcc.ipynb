# Step 5: Preprocessing
Typical preprocessing steps include normalization, filtering, and scaling:

```python
import scanpy as sc

# Step 1: Load the data
adata_normal = sc.read_10x_mtx('E:/Projects/CADD/Single cell Retenoic/Normal/', var_names='gene_symbols', cache=True)
adata_tumor = sc.read_10x_mtx('E:/Projects/CADD/Single cell Retenoic/Tumor/', var_names='gene_symbols', cache=True)

# Step 2: Concatenate the datasets (normal and tumor)
adata_combined = adata_normal.concatenate(adata_tumor, batch_key='sample_type')

# Filter mitochondrial genes (ensure this only affects genes, not cells)
adata_combined = adata_combined[:, ~adata_combined.var_names.str.startswith('MT-')]

# Proceed with the rest of the preprocessing steps
# Step 1: Basic filtering - Remove genes that are detected in fewer than 3 cells
sc.pp.filter_genes(adata_combined, min_cells=3)

# Step 2: Filtering cells - Remove cells with low or high total counts, or too high mitochondrial gene expression
sc.pp.filter_cells(adata_combined, min_genes=200)

# Step 3: Calculate the percentage of mitochondrial genes for each cell (optional for further filtering)
adata_combined.obs['percent_mt'] = (adata_combined[:, adata_combined.var_names.str.startswith('MT-')].X.sum(axis=1).A1 /
                                    adata_combined.X.sum(axis=1).A1) * 100

# Optionally, filter out cells with more than 5% mitochondrial genes (adjust threshold as needed)
adata_combined = adata_combined[adata_combined.obs['percent_mt'] < 5, :]

# Step 4: Normalize and scale the data (recommended for clustering)
sc.pp.normalize_total(adata_combined, target_sum=1e4)  # Normalize to a total count of 10,000 per cell
sc.pp.log1p(adata_combined)  # Log-transform the data
sc.pp.scale(adata_combined, max_value=10)  # Scale each gene to have zero mean and unit variance

# Step 5: Compute the neighborhood graph
sc.pp.neighbors(adata_combined, n_neighbors=10, n_pcs=40)

# Step 6: Cluster the cells using Leiden algorithm
sc.tl.leiden(adata_combined)

# Step 7: Compute UMAP for visualization
sc.tl.umap(adata_combined)

# Step 8: Plot UMAP with the sample type and Leiden clustering
sc.pl.umap(adata_combined, color=['sample_type', 'leiden'])
```

# Step 8: Differential Expression Analysis
Identify genes that are differentially expressed between Normal and Tumor samples:

# Y-Axis (Scores):
The y-axis represents the statistical scores from the differential expression analysis. These scores quantify how strongly a gene is differentially expressed between the groups.

ex : On the y-axis, the number 100 represents the statistical score of a gene's differential expression between the two groups ("Normal" and "Tumor").

# the X-Axis
The x-axis in the plot represents the genes that were found to be the most important (or significant) for distinguishing between the two groups: "Normal" and "Tumor".

ex : On the x-axis, the number 15 represents the 15th ranked gene based on its differential expression score.

```python
sc.tl.rank_genes_groups(adata_combined, 'sample_type', method='t-test')
sc.pl.rank_genes_groups(adata_combined, n_genes=20, sharey=False)
```

```python
# Get the ranked genes for "Normal" and "Tumor"
ranked_genes = adata_combined.uns['rank_genes_groups']
df_genes = pd.DataFrame({
    group: ranked_genes['names'][group] 
    for group in ranked_genes['names'].dtype.names
})

print(df_genes)
```

# Identify Top Differentially Expressed Genes in Each Cluster
You can use the rank_genes_groups method to identify the most highly differentially expressed genes for each cluster. These genes often give clues about the biological identity of the cluster.

```python
# Compute the top differentially expressed genes for each cluster
sc.tl.rank_genes_groups(adata_combined, 'leiden', method='t-test')

# Get the names of the top genes for each cluster
top_genes = adata_combined.uns['rank_genes_groups']['names']

# Print the top 10 genes for each cluster
for cluster in top_genes.dtype.names:
    print(f"Top genes for Cluster {cluster}:")
    print(top_genes[cluster][:10])  # Show top 10 genes for this cluster
    print()
```
